# Meat consumption in the world

```python
#read the libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
""" Created a functions for line and bar and pie plots
in line_p funtion we are ploting the line graph """
# Order of magnitude or value.
def convert_to_mln(data):
    return data['Value'].div(10**3).round(2)
# Choose years, when was eaten meat
def time_consumption(data,year_begin, year_end):
    result = (data['TIME'] >= year_begin) & (data['TIME']  < year_end)
    return data.loc[result]
# creating function to create a bar graph
def bar_plot(xdata,ydata,style,title,xsize,ysize,ylabel,xlabel):

    sns.set_style(style)
    plt.figure(figsize=(xsize,ysize))
    plt.title(title)
    sns.barplot(x=xdata,y=ydata)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.savefig("bar-plot.png")
    return plt.show()
```

```python
# creating function to create a line graph
def line_p(x_axis, y_axis, name, col):
    
    return plt.plot(x_axis,y_axis, '.-',label= name, color= col)
```

```python
def get_meat_consumption(filename):


    # Calculate total meat consumption by subject and sort by descending order
    meat = filename.groupby(by='SUBJECT')['Value'].sum().reset_index().sort_values('Value', ascending=False)

    # Extract labels and values for the pie chart
    labels = meat['SUBJECT']
    values = meat['Value']

    # Create the pie chart
    plt.figure(figsize=(10,10))
    plt.pie(values, labels=labels, autopct='%1.2f%%', startangle=0, explode=(0.1, 0.04, 0.01, 0.01))
    plt.title('Total Meat Consumption Worldwide')

    # Save the plot to a file
    plt.savefig('meat_consumption_pie.png')

    # Display the plot
    plt.show()

    # Return the meat DataFrame
    return meat






    
```

```python
#read the dataset
meat_consumption=pd.read_csv("meat_consumption_worldwide.csv")
print(meat_consumption)
#filter the given data 
meat_consumption.head(5)
```

```python
meat_data_tonne = meat_consumption[meat_consumption['MEASURE']=='THND_TONNE']

meat_data_tonne = meat_data_tonne.loc[meat_data_tonne['LOCATION']=='WLD']
meat_data_tonne["TOTAL_IN_MLN"] = convert_to_mln(meat_data_tonne)
meat_data_tonne= time_consumption(meat_data_tonne,1992, 2019)[['TIME','TOTAL_IN_MLN']]
meat_data_group_tonne = meat_data_tonne.groupby(['TIME'])['TOTAL_IN_MLN'].agg(mean='mean').reset_index()
pd.DataFrame(meat_data_group_tonne['mean'].describe().round(2))
```

```python
bar_plot(meat_data_group_tonne['TIME'],meat_data_group_tonne['mean'],'whitegrid','Average distributions of meat consumption in years 1992-2018',20,5,'Avarage consumptions [mln tonne]','Year')
```

```python
#filtering data as requied
meat_consumption=meat_consumption[(meat_consumption["LOCATION"] == "AUS") & (meat_consumption["MEASURE"] == "KG_CAP")
          & (meat_consumption["TIME"]>= 2010) & (meat_consumption["TIME"]<= 2022)]
```

```python
meat_sheep = meat_consumption[(meat_consumption["SUBJECT"]=="SHEEP")]
meat_beef = meat_consumption[(meat_consumption["SUBJECT"]=="BEEF")] #sub dataframe for Beef
meat_poultry = meat_consumption[(meat_consumption["SUBJECT"]=="POULTRY")] #sub dataframe for Poultry
meat_pig = meat_consumption[(meat_consumption["SUBJECT"]=="PIG")] ##sub dataframe for Pig
```

```python
#produsing a line graph  using a line_p function
#line_p(meat_sheep.TIME, meat_sheep.Value, "Sheep","blue")
plt.figure(figsize=(10, 8))
line_p(meat_beef.TIME, meat_beef.Value, "Beef","red")
line_p(meat_poultry.TIME, meat_poultry.Value, "Poultry","green")
line_p(meat_pig.TIME, meat_pig.Value, "Pig","yellow")
line_p(meat_sheep.TIME, meat_sheep.Value, "Sheep","blue")
plt.xlabel("Years", size=12)
plt.ylabel("Tons", size=12)
plt.title("Meat consumption of Austrilla from 2010 to 2022")
plt.legend()
plt.savefig("line-plot.png")
plt.show()
```
